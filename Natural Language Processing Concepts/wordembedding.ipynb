# Steps

### 1] Initialize Sentences.
### 2] one hot representation of sentences  --> return index from the dictionary.
### 3] one hot representation. --> embedding layer keras to form embedding matrix.
### 4] Embedding Matrix

# Step: import libraries

```python
from tensorflow.keras.preprocessing.text import one_hot
```

# Step1 : Initializing Sentence

```python
sent = ['the glass of milk',
       'the glass of juice',
       'The cup of tea',
       'I love machine learning',
       'I love deep learning',]
```

# step2: one hot representation of sentences

```python
# vocabulary size
voc_size = 10000
```

```python
one_hot_rep = [one_hot(word,voc_size)for word in sent]
```

```python
one_hot_rep
```

# step 3: Word embedding representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_len = 8
```

#### all sentences should have same length . thats why we use padding

```python
embedded_docs = pad_sequences(one_hot_rep,padding='pre',maxlen = sent_len)
```

```python
embedded_docs
```

```python
dimension = 5
```

# step4: Adding embedding layer

```python
model = Sequential()
```

```python
model.add(Embedding(voc_size,10,input_length = sent_len))
```

```python
model.compile('adam','mse')  #adam optimizer and mean square
```

```python
model.summary()
```

```python
model.predict(embedded_docs) #
```
