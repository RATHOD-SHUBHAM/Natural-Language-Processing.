```python
import pandas as pd
```

# Step1 :  Read the data

```python
df = pd.read_csv('Stock_Dataa.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

# Step 2: Split train and test data

```python
train = df[df['Date'] < '20150101']
```

```python
train.head()
```

```python
test = df[df['Date']>'20141231']
```

```python
test.head()
```

# : Data Preprocessing

# Step3: Remove Punctuation 

```python
data = train.iloc[:,2:27]
```

```python
data.head()
```

```python
data.replace("[^A-Aa-z]"," ",regex = True,inplace = True)
```

```python
data
```

```python
# For making things easy lets change the column name from top1 top2 to just 1 2 3..
```

```python
list1 = [i for i in range(25)]
```

```python
#list1
# list1 contains list of integer
```

```python
new_index = [str(i) for i in list1]
```

```python
#new_index
# new_index contain list of string
```

```python
# make the new_index as column name.
data.columns = new_index
```

```python
data.head()
```

# Step 4: Convert all the upper case to lower case value

```python
# data['0']
```

```python
for index in new_index:
    data[index] = data[index].str.lower()
```

```python
data.head()
```

# Step 5: for each row: Combine all the column values into a single paragraph

```python
headlines = []
```

```python
# Gives the row 
data.index
```

```python
for rows in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[rows,0:25]))
```

```python
headlines[0]
```

# step 6: Perform Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
countvector = CountVectorizer(ngram_range = (2,2))
```

```python
traindataset = countvector.fit_transform(headlines)
```

# step 7: use ML algorithm

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randomclassifier = RandomForestClassifier(n_estimators = 200,criterion='entropy')
```

```python
randomclassifier.fit(traindataset,train['Label'])
```

# step 8:  Perform Same things on test data set

```python
# we wont pre process the data
```

```python
test_transform = []
```

```python
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
```

```python
testdataset = countvector.transform(test_transform)
```

### We trained our ML model with train data then used to make predictions using test data

```python
predictions = randomclassifier.predict(testdataset)
```

# Step 9: Confusion matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
matrix = confusion_matrix(test['Label'],predictions)
```

```python
print(matrix)
```

# Step 10: compute accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(test['Label'],predictions)
```

```python
print(score)
```

# step 11: Classification report

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(test['Label'],predictions)
```

```python
print(report)
```
